# Análise do Eleitorado


```python
import pandas as pd
```

```python
#Carregando arquivo csv
df = pd.read_csv('eleitorado_municipio_2018.csv',sep=';',encoding='latin1')

df.head()





```

```python
df.info()
```

```python
# As cinco didades com eleitores deficientes
df.nlargest(5,'QTD_ELEITORES_DEFICIENTE')
```

```python
print('Total Eleitores:', df['QTD_ELEITORES'].sum())
print('Eleitoras:', df['QTD_ELEITORES_FEMININO'].sum())
print('Eleitores:', df['QTD_ELEITORES_MASCULINO'].sum())
print('Não Informados:', df['QTD_ELEITORES_NAOINFORMADO'].sum())

```

```python
# criar variaveis para porcentagens
total_eleitores = df['QTD_ELEITORES'].sum()
tot_fem = df['QTD_ELEITORES_FEMININO'].sum()
tot_nao =  df['QTD_ELEITORES_NAOINFORMADO'].sum()
tot_masc = df['QTD_ELEITORES_MASCULINO'].sum()
```

```python
# Porcentagem Eleitorado    
print('Eleitoras',((tot_fem/total_eleitores)*100).round(2) )  
print('Eleitores',((tot_masc/total_eleitores)*100).round(2) )  
print('Não',((tot_nao/total_eleitores)*100).round(2) )  


```

```python
# QUANTOS municipios com mais homes que mulheres
df[df['QTD_ELEITORES_MASCULINO'] > df['QTD_ELEITORES_FEMININO']].count()
```

```python
# criar nova coluna para mostrar relação fem?masc
df['RELAÇÃO']= df['QTD_ELEITORES_FEMININO']/df['QTD_ELEITORES_MASCULINO']
```

```python
# MUNICIPIOS COM MAIOR RELAÇÃO
df.nlargest(5,'RELAÇÃO').T
```

```python
# criano um novo dataframe somente com municipios do brasil.
df_brasil = df[df['CD_PAIS']==1].copy()

df_brasil.head(3).T
```

```python
# MUNICIPIOS COM MAIOR RELAÇÃO
df_brasil.nlargest(5,'RELAÇÃO').T
```

```python
#municipios com a menor relação f/m

df_brasil.nsmallest(5,'RELAÇÃO').T
```

```python
# pedindo ára mostrar o grafico
%matplotlib inline
# plotar grafico da relação m/f
df_brasil['RELAÇÃO'].plot.hist(bins=30)
```

```python
# fazendo grafico com outra biblioteca
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# plotar a relação f/m
sns.distplot(df_brasil['RELAÇÃO'], bins=50, color='red')
```

```python
# plotar a relação f/m, sem linha deccontorno
sns.distplot(df_brasil['RELAÇÃO'], bins=50, color='red', kde=False)
```

```python
# plotar a relação f/m, sem linha deccontorno
sns.distplot(df_brasil['RELAÇÃO'], bins=50, color='red', kde=False)
# embelezando o grafico
plt.title('Relação Eleitores/Eleitoras', fontsize=18)
plt.xlabel('Eleitores/Eleitoras', fontsize=14)
plt.ylabel('Frequência', fontsize=14)
plt.axvline(1.0, color='green', linestyle='--')
```

```python
# verificar a distribuição mostrando os ponto dos municipios


sns.swarmplot(data=df_brasil, x= 'NM_REGIAO', y='RELAÇÃO')
```

```python
# Desenha linha horizontal


sns.swarmplot(data=df_brasil, x= 'NM_REGIAO', y='RELAÇÃO')
plt.axhline(1.0, color='green', linestyle='--')
```
